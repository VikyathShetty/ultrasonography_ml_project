```python
#it will drop the colums which has nan value greater than 20 percent
perc = 50.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
#df.shape[0] - tells the no of rows
df = df.dropna( axis=1, thresh=min_count)
```

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score,StratifiedKFold,train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#we split X and y to remove the data leakage
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size = 0.2,stratify=y)
```

```python
X_train.corr()
```

```python
#Removing multicollinearity
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train,0.7)
len(set(corr_features))
```

```python
X_train = X_train.drop(corr_features,axis=1)
X_test = X_test.drop(corr_features,axis=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#MinMmax scaler
mm = MinMaxScaler()
X_t1 = mm.fit_transform(X_train)
X_t2 = mm.transform(X_test)
```

```python
#VariacnceThreshold
vt = VarianceThreshold(threshold = 0.01)
vt.fit_transform(X_t1).shape
```

```python
vt.transform(X_t2).shape
```

```python
pipe = Pipeline([
    ("s2",MinMaxScaler()),
    ("s3",VarianceThreshold(threshold=0.01)),
    ("s4",RandomForestClassifier(random_state=0))
    
])
```

```python
#Display Pipeline
from sklearn import set_config
set_config(display = "diagram")
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps
```

```python
#predicted value
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
score = cross_val_score(pipe,X_train,y_train,n_jobs=-1,cv = StratifiedKFold(random_state = 0,shuffle = True))
score.mean()
```

```python
#cross validation in pipeline
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring = "accuracy").mean()
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
